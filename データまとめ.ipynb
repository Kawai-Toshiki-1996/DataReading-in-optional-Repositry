```python
import os
import pandas as pd
import csv
import numpy as np
os.getcwd()
```

```python
f=open("CRANE20180112.csv","w",encoding="UTF-8")
f.write("Year,Month,Day,Time,Unknown,Latitude,Longi(Deg),Longitude,H,T,MeanDirection,WindDirection,Wind,YesorNo,Numer")
f.write('\n')

SeaState = pd.read_csv('sealist.csv', encoding='shift_jis',engine='python')
SeaList=SeaState.values

SeaNum=0
csv_list=os.listdir()
for i in os.listdir()[1:len(os.listdir())-3]:#range(1,len(os.listdir())-4+1)
    RowNum, GuideNum=0,0
    csvfile=open(i)#os.listdir()[i]
    print(csvfile)
    for row in  csv.reader(csvfile):
        
        if RowNum == GuideNum :
            print(str(row[0])+" : "+str(SeaList[SeaNum][0:5]))
            if SeaList[SeaNum][13] == 1 :#
                SeaNum+=1#
                GuideNum+=3#csv3

            elif SeaList[SeaNum][13] == 0 :#     
                [ f.write(str(SeaList[SeaNum][j])+",") for j in range(15)]
                SeaNum+=1#
        
        elif (RowNum-26) == GuideNum:
            [ f.write(str(row[j])+",") for j in range(1,37)]
            f.write('\n')
            GuideNum+=28
        
        elif (RowNum-2) >= GuideNum :
            #print(str(RowNum)+" : "+str(np.shape(row)))
            [ f.write(str(row[j])+",") for j in range(1,37)]
        
        RowNum+=1

f.close()
print("End")    
```
